```python
iron=df.groupby('Category')['Iron (% Daily Value)'].mean()
calcium=df.groupby('Category')['Calcium (% Daily Value)'].mean()
```

```python
category = iron.index #putting the index name in category
Iron = iron.values  #putting the values of iron in Iron
Calcium = calcium.values
```

```python
plt.figure(figsize=(10, 4), dpi=100)


X_axis = np.arange(len(category))

plt.bar(X_axis -0.2, Iron,0.4, label = 'Iron', color = 'red')
plt.bar(X_axis + 0.2, Calcium, 0.4, label = 'Calcium', color = 'yellow')

plt.xticks(X_axis, category)
plt.xlabel("Menu ")
plt.title("Distribution of Iron and Calcium in McDonald's Foods")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# Summary 

- Coffee and tea have the highest saturated fat to cholesterol ratio among all categories.




- Salads have the highest vitamin A and vitamin C content among all categories.

- Beef and pork have the highest iron content among all categories.

- Smoothies and shakes have the highest calcium content among all categories.

* The analysis reveals some insights into the nutritional value and health implications of different categories and items.
* The analysis also suggests some possible areas for improvement or further investigation, such as reducing the trans fat and sodium content, increasing the dietary fiber and protein content, and diversifying the vitamin and mineral sources.
